```python
# First and last index of an ele in sorted array

def find(arr,n,x):
    first = -1
    last = -1
    f = False
    low = 0
    high = n-1
    while(low<=high):
        mid = low+(high-low)//2
        if arr[mid]==x:
            high=mid-1
            f = True
        elif arr[mid]<x:
            low = mid+1
        else:
            high = mid-1
    if not f:
        return [-1,-1]
    first = high+1
    low = 0
    high = n-1
    while(low<=high):
        mid = low+(high-low)//2
        if arr[mid]==x:
            low = mid+1
        elif arr[mid]<x:
            low = mid+1
        else:
            high = mid-1
    last = low-1
    return [first,last]

arr = list(map(int,input().split()))
x = int(input())
n = len(arr)
ans = find(arr,n,x)
print(*ans,sep=" ")
```

```python
# Count number of occurrences (or frequency) in a sorted array

def count(arr,n,x):
    first = -1
    last = -1
    f = False
    low = 0
    high = n-1
    while(low<=high):
        mid = low+(high-low)//2
        if arr[mid]==x:
            high=mid-1
            f = True
        elif arr[mid]<x:
            low = mid+1
        else:
            high = mid-1
    if not f:
        return 0
    first = high+1
    low = 0
    high = n-1
    while(low<=high):
        mid = low+(high-low)//2
        if arr[mid]==x:
            low = mid+1
        elif arr[mid]<x:
            low = mid+1
        else:
            high = mid-1
    last = low-1
    ans = last-first+1
    return ans

arr = list(map(int,input().split()))
x = int(input())
n = len(arr)
ans = count(arr,n,x)
print(ans)
```

```python
# Number of times a sorted array is rotated

def no(arr):
#     if len(arr)==1:
#         return 0
    if arr[0]<=arr[-1]:
        return 0
    n = len(arr)
    low = 0
    high = len(arr)-1
    while(low<=high):
        if arr[low]<=arr[high]:
            return low
        mid = low+(high-low)//2
        prev = (mid-1+n)%n
        next = (mid+1)%n
        if arr[mid]<=arr[prev] and arr[mid]<=arr[next]:
            return mid
        elif arr[low]<=arr[mid]:
            low = mid+1
        elif arr[mid]<=arr[high]:
            high = mid-1

arr = list(map(int,input().split()))
print(no(arr))
```

```python
# Search in an nearly sorted array

def nearly(arr,x):
    n = len(arr)
    low = 0
    high = n-1
    while(low<=high):
        mid = low+(high-low)//2
        if arr[mid]==x:
            return mid
        elif mid-1>=0 and arr[mid-1]==x:
            return mid-1
        elif mid+1<=n-1 and arr[mid+1]==x:
            return mid+1
        elif arr[mid]<x:
            low = mid+2
        else:
            high = mid-2
    return -1

arr = list(map(int,input().split()))
x = int(input())
print(nearly(arr,x))
```

```python
# Floor of an element in a Sorted Array

def floor(arr,x):
    n = len(arr)
    low = 0
    high = n-1
    while(low<=high):
        mid = low+(high-low)//2
        if arr[mid]==x:
            return x
        elif arr[mid]<x:
            low = mid+1
        else:
            high = mid-1
    if high!=-1:
        return arr[high]
    return -1
    
arr = list(map(int,input().split()))
x = int(input())
print(floor(arr,x))
```

```python
# Ceil of an element in a Sorted Array

def ceil(arr,x):
    n = len(arr)
    low = 0
    high = n-1
    while(low<=high):
        mid = low+(high-low)//2
        if arr[mid]==x:
            return x
        elif arr[mid]<x:
            low = mid+1
        else:
            high = mid-1
    if low!=n:
        return arr[low]
    return -1
    
arr = list(map(int,input().split()))
x = int(input())
print(ceil(arr,x))
```

```python
# Find position of an element in a sorted array of infinite numbers
# T.C. - O(Log p), p is position of element to be searched

def inf_search(arr,x):
    n = len(arr)
    low = 0
    high = 1
    while(arr[high]<x):
        low = high
        high = high*2
        if high>=n:       # Extra condition
            high = n-1
            break
    while(low<=high):
        mid = low+(high-low)//2
        if arr[mid]==x:
            return mid
        elif arr[mid]<x:
            low = mid+1
        else:
            high = mid-1
    return -1
    
arr = list(map(int,input().split()))
x = int(input())
print(inf_search(arr,x))
```

```python
# Search in a row wise and column wise sorted matrix
def search_mat(mat,x,n):
    i = 0
    j = n-1
    while(i>=0 and i<n and j>=0 and j<n):
        if mat[i][j]==x:
            return [i,j]
        elif mat[i][j]>x:
            j-=1
        else:
            i+=1
    return -1
    
n = int(input())
mat = []
for i in range(n):
    k = list(map(int,input().split(",")))
    mat.append(k)
x = int(input())
print(search_mat(mat,x,n))
```

```python
# Square root of a number

def sq_root_no(n):              ### T.C.:- O(sqrt(n))
    i = 1
    while(i*i<=n):
        i = i+1
    return i-1

def sq_root(n):                 ### T.C.:- O(logn)(Using binary search)
    low = 1
    high = n
    ans = 0
    while(low<=high):
        mid = low+(high-low)//2
        if mid*mid<n:
            low = mid+1
            ans = mid
        elif mid*mid>n:
            high = mid-1
        else:
            ans = mid
            break
    return ans
    
n = int(input())
print(sq_root_no(n))
```

```python
# Single Element In Sorted Array(logn sol for XOR ques)

def sing(arr):                       ### T.C.:- O(logN)
    low = 0                          ### S.C.:- O(1)
    high = len(arr)-2                # Single element can be in the last............
    while(low<=high):
        mid = (low+high)>>1
        if arr[mid]==arr[mid^1]:
            low = mid+1
        else:
            high = mid-1
    return arr[low]                 # Breakpoint is met at low(low stands imm. after left half)
    
arr = list(map(int,input().split()))
print(sing(arr))
```

```python
# Median of 2 sorted arrays

# Naive: Apply merge function on arr1 and arr2. Get median(T.C.:- O(n1+n2), S.C.:- O(n1+n2)
#        Apply merge function and keep cnt for middle ele.Then compute and return median(T.C.:- O(n1+n2), S.C.:- O(1))

import sys
def arr_med(arr1,arr2):
    if len(arr2)<len(arr1):                  ### T.C.:- O(log(min(n1,n2)))
        return arr_med(arr2,arr1)            ### S.C.:- O(1)
    n1 = len(arr1)
    n2 = len(arr2)
    low = 0
    high = n1
    while(low<=high):
        cut1 = (low+high)>>1
        cut2 = ((n1+n2+1)//2)-cut1
        l1 = arr1[cut1-1] if cut1 else -sys.maxsize
        l2 = arr2[cut2-1] if cut2 else -sys.maxsize
        r1 = arr1[cut1] if cut1!=n1 else sys.maxsize
        r2 = arr2[cut2] if cut2!=n2 else sys.maxsize
        if l1<=r2 and l2<=r1:
            if (n1+n2)%2:
                ans = max(l1,l2)
                return ans
            else:
                k1 = max(l1,l2)
                k2 = min(r1,r2)
                ans = (k1+k2)/2
                return ans
        elif l1>r2:
            high = cut1-1
        else:
            low = cut1+1
    return 0.0
    
arr1 = list(map(int,input().split()))
arr2 = list(map(int,input().split()))
print(arr_med(arr1,arr2))
```
