```python
# Implement stack using array
class Stack:
    def __init__(self,n):
        self.size = n
        self.arr = [None]*n
        self.top = -1
    
    def push(self,x):                         # Method 2................(Using top pointer)
        if self.top<self.size-1:
            self.top+=1
            self.arr[self.top] = x
        else:
            print("Stack Overflow")
    
    def pop(self):
        if self.top==-1:
            print("Stack Underflow")
        else:
            x = self.arr[self.top]
            self.top-=1
            return x
    
    def isEmpty(self):
        if self.top==-1:
            return True
        return False
    
    def peek(self):
        if self.top==-1:
            print('None')
            return
        else:
            return self.arr[self.top]
    
    def size_st(self):
        k = self.top+1
        return k
    
s = Stack(5)
s.peek()
print(s.size_st())
s.pop()
print(s.isEmpty())
s.push(10)
print(s.size_st())
s.push(20)
s.push(30)
print(s.pop())
print(s.peek())
print(s.size_st())
```

```python
# Implement stack using Linked List
class Stack:
    def __init__(self):
        self.head = None
        
    def isEmpty(self):
        temp = self.head
        if temp is None:
            return True
        return False
        
    def push(self,x):
        new_n = StackNode(x)
        new_n.next = self.head
        self.head = new_n
        print(x,"pushed to stack")
        
    def pop(self):
        if(self.isEmpty()):
            print("Stack Underflow")
            return
        else:
            temp = self.head
            self.head = temp.next
            print(temp.data,"popped out of stack")
            temp = None
    
    def peek(self):
        if(self.isEmpty()):
#             print("None")
            return
        else:
            temp = self.head.data
            return temp
    
    def size(self):
        count = 0
        temp = self.head
        while(temp):
            count+=1
            temp = temp.next
        return count
    
        
class StackNode:
    def __init__(self,data):
        self.data = data
        self.next = None
        
s = Stack()
s.pop()
s.push(10)
s.push(20)
s.push(30)
s.pop()
print(s.isEmpty())
print(s.size())
print(s.peek())
```

```python
# Nearest Greater to Right
def ngr(arr):
    st = []
    rev = []
    for i in range(len(arr)-1,-1,-1):
        if len(st)==0:
            rev.append(-1)
        elif len(st)>0 and st[-1]>arr[i]:
            rev.append(st[-1])
        else:
            while(len(st)>0 and st[-1]<=arr[i]):
                st.pop()
            if len(st)==0:
                rev.append(-1)
            else:
                rev.append(st[-1])
        st.append(arr[i])
    ans = rev[::-1]
    return ans
    
arr = list(map(int,input().split()))
ans = ngr(arr)
print(*ans,sep=" ")
```

```python
# Stock Span Problem
def stock_span(arr):  # NGL variation
    st = []
    ans = []
    for i in range(len(arr)):
        if len(st)==0:
            ans.append(-1)
        elif len(st)>0 and arr[st[-1]]>arr[i]:
            ans.append(st[-1])
        else:
            while(len(st)>0 and arr[st[-1]]<=arr[i]):
                st.pop()
            if len(st)==0:
                ans.append(-1)
            else:
                ans.append(st[-1])
        st.append(i)
    for i in range(len(arr)):
        ans[i] = i-ans[i]
    return ans
    
arr = list(map(int,input().split()))
ans = stock_span(arr)
print(*ans,sep=" ")
```

```python
# Maximum Area Histogram
def max_ar_his(arr):   # NSL and NSR variation
    st = []
    left = []
    right = []
    n = len(arr)
    for i in range(n):
        while(len(st)>0 and arr[st[-1]]>=arr[i]):
            st.pop()
        if len(st)==0:
            left.append(-1)
        else:
            left.append(st[-1])
        st.append(i)
    st = []
    for i in range(n-1,-1,-1):
        while(len(st)>0 and arr[st[-1]]>=arr[i]):
            st.pop()
        if len(st)==0:
            right.append(n)
        else:
            right.append(st[-1])
        st.append(i)
    right = right[::-1]
    width = [0]*n
    area = 0
    for i in range(n):
        width[i] = right[i]-left[i]-1
        area = max(area,arr[i]*width[i])
    return area
    
arr = list(map(int,input().split()))
print(max_ar_his(arr))
```

```python
# Max. Area Rectangle in binary matrix
def mah(arr):
    st = []
    left = []
    right = []
    l = len(arr)
    for i in range(l):
        while(st and arr[st[-1]]>=arr[i]):
            st.pop()
        if not st:
            left.append(-1)
        else:
            left.append(st[-1])
        st.append(i)
    st = []
    for i in range(l-1,-1,-1):
        while(st and arr[st[-1]]>=arr[i]):
            st.pop()
        if not st:
            right.append(l)
        else:
            right.append(st[-1])
        st.append(i)
    right = right[::-1]
    area = 0
    for i in range(l):
        area = max(area,arr[i]*(right[i]-left[i]-1))
    return area

def mar_bin_matrix(matrix):
    n = len(matrix)
    m = len(matrix[0])
    arr = [0]*m
    for i in range(m):
        arr[i] = matrix[0][i]
    ans = mah(arr)
    for i in range(1,n):
        for j in range(m):
            if matrix[i][j]==0:
                arr[j] = 0
            else:
                arr[j]+=matrix[i][j]
        ans = max(ans,mah(arr))
    return ans
    
n = int(input())
matrix = []
for _ in range(n):
    matrix.append(list(map(int,input().split())))
print(mar_bin_matrix(matrix))
```

```python
# Trapping Rain Water
def trapping_water(arr):
    n = len(arr)
    left = [0]*n
    right = [0]*n
    left[0] = arr[0]
    right[n-1] = arr[n-1]
    for i in range(1,n):
        left[i] = max(left[i-1],arr[i])
    for i in range(n-2,-1,-1):
        right[i] = max(right[i+1],arr[i])
    ans = 0
    for i in range(n):
        ans+=min(left[i],right[i])-arr[i]
    return ans
    
arr = list(map(int,input().split()))
print(trapping_water(arr))
```

```python
# Implement Queue Using Arrays

class Queue:
    def __init__(self,n):
        self.size = n
        self.front = 0
        self.rear = 0
        self.q = [None]*n
        self.cnt = 0
        
    def push(self,x):
        if self.cnt==self.size:
            print("Queue Full")
        else:
            self.q[self.rear%self.size] = x
            self.rear+=1
            self.cnt+=1
        
    def pop(self):
        if self.front==self.rear:
            print("Queue Empty")
        else:
            self.q[self.front%self.size] = -1
            self.front+=1
            self.cnt-=1
    
    def peek(self):
        if self.front==self.rear:
            return None
        else:
            k = self.q[self.front%self.size]
            return k
        
    def size(self):
        return self.cnt
    
    def isEmpty(self):
        return self.cnt==0
    
Q = Queue(5)
print(Q.isEmpty())
Q.push(1)
Q.push(2)
Q.push(3)
Q.pop()
Q.push(4)
print(Q.peek())
print(Q.isEmpty())
Q.push(10)
Q.push(7)
Q.push(9)
Q.push(8)
Q.pop()
print(Q.peek())
```

```python
# Implement Stack using Queues(Using 2 queues)
# We keep elements in q1
# Other method: Making pop operation costly. Dequeue everything from q1 except one ele and enqueue to q2. Pop or top(delete and 
# add)it as required. Swap(T.C.:- O(N), S.C.:- O(2N)~O(N))(Both pop and top are modified.)

class Stack:
    
    def __init__(self):
        self.q1 = []
        self.q2 = []
        
    def push(self,x):                            # Making push operation costly
        self.q2.append(x)                        ### T.C.:- O(N)
        while(self.q1):                          ### S.C.:- O(2N)~O(N)
            self.q2.append(self.q1[0])
            self.q1.pop(0)
        temp = self.q1
        self.q1 = self.q2
        self.q2 = temp
        
    def pop(self):                               # Remaining all op. O(1)
        if self.q1:
            return self.q1.pop(0)
        return -1
    
    def top(self):
        if self.q1:
            return self.q1[0]
        return -1
    
    def size(self):
        return len(self.q1)
    
s = Stack()
s.push(1)
s.push(2)
s.push(3)
print(s.size())
print(s.pop())
print(s.top())
print(s.pop())
print(s.pop())
print(s.pop())
print(s.top())
print(s.size())
```

```python
# Implement Stack using Queues(Using single queue)
# Recently added -> front of queue

class Stack:
    
    def __init__(self):
        self.q1 = []
        
    def push(self,x):                    ### T.C.:- O(N)
        self.q1.append(x)                ### S.C.:- O(N)
        k = len(self.q1)
        while(k-1):
            i = self.q1.pop(0)
            self.q1.append(i)
            k-=1
    
    def pop(self):                       # Remaining all op. O(1)
        if self.q1:
            self.q1.pop(0)
            
    def top(self):
        if self.q1:
            return self.q1[0]
        return -1
    
    def size(self):
        return len(self.q1)
    
s = Stack()
s.push(1)
s.push(2)
s.push(3)
print(s.size())
print(s.top())
s.pop()
print(s.top())
s.pop()
print(s.top())
print(s.size())
s.pop()
print(s.size())
```

```python
# Sliding Window Maximum

# Naive: for i->0 to n+k-1                         T.C.:- O(N*k)
#          mx = a[i]                               S.C.:- O(1)
#          for j->i to i+k
#            mx = max(mx,a[j])
from collections import deque

def slide_max(arr,k):                          ### T.C.:- O(N)+O(N) ~~ O(N)(Amortised)
    q = deque()                                ### S.C.:- O(k)
    ans = []
    for i in range(len(arr)):                  # Storing ele in q in decreasing order
        if(q and q[0]==i-k):
            q.popleft()
        while(q and arr[q[-1]]<=arr[i]):
            q.pop()
        q.append(i)
        if(i>=k-1):
            ans.append(arr[q[0]])
    return ans
    
arr = list(map(int,input().split()))
k = int(input())
ans = slide_max(arr,k)
print(*ans,sep=" ")
```

```python
# Find maximum of minimums of every window size

# Naive: Iterate for all window size from 1 to n,then,O(n*k) method for finding maximum of min of particular window size
# T.C.:- O(N^3), S.C.:- O(1)

def max_min(arr):                                        ### T.C.:- O(N)
    n = len(arr)                                         ### S.C.:- O(N)
    st = []
    left = []
    right = []
    for i in range(n):
        while(st and arr[st[-1]]>=arr[i]):
            st.pop()
        if not st:
            left.append(-1)
        else:
            left.append(st[-1])
        st.append(i)
    st = []
    for i in range(n-1,-1,-1):
        while(st and arr[st[-1]]>=arr[i]):
            st.pop()
        if not st:
            right.append(n)
        else:
            right.append(st[-1])
        st.append(i)
    right = right[::-1]
    ans = [0]*(n+1)
    for i in range(n):
        l = right[i]-left[i]-1
        ans[l] = max(ans[l],arr[i])
    for i in range(n-1,0,-1):
        ans[i] = max(ans[i],ans[i+1])
    return ans
    
arr = list(map(int,input().split()))
print(max_min(arr))
```

```python
# Implement Min Stack(With O(N) Extra Space, T.C.:- O(1) for every operation)

class MinStack:
    
    def __init__(self):
        self.st = []
        self.ss = []
        
    def push(self,x):
        self.st.append(x)
        if not self.ss or x<=self.ss[-1]:
            self.ss.append(x)
    
    def pop(self):
        if not self.st:
            return -1
        k = self.st.pop()
        if k==self.ss[-1]:
            self.ss.pop()
        return k
    
    def top(self):
        if not self.st:
            return -1
        return self.st[-1]
    
    def getMin(self):
        if not self.ss:
            return -1
        return self.ss[-1]
    
s = MinStack()
print(s.pop())
print(s.top())
print(s.getMin())
s.push(10)
s.push(2)
s.push(3)
print(s.getMin())
print(s.top())
s.push(1)
print(s.getMin())
s.pop()
print(s.getMin())
s.pop()
s.pop()
print(s.top())
print(s.getMin())
```

```python
# Implement Min Stack(With O(1) Extra Space, T.C.:- O(1) for every operation)

class MinStack:
    
    def __init__(self):
        self.st = []
        self.minEle = -1
    
    def push(self,x):
        if not self.st:
            self.st.append(x)
            self.minEle = x
        else:
            if x>=self.minEle:
                self.st.append(x)
            else:
                k = 2*x-self.minEle                     # (2x-ME)
                self.st.append(k)                       # Flag value is stored in stack
                self.minEle = x                         # Actual min in minEle
    
    def pop(self):
        if not self.st:
            return -1
        else:
            if self.st[-1]>=self.minEle:
                self.st.pop()
            else:
                self.minEle = 2*self.minEle-self.st[-1]        # (2ME-stack[top]) To get prev min.
                self.st.pop()
                
    def top(self):
        if not self.st:
            return -1
        else:
            if self.st[-1]>=self.minEle:
                return self.st[-1]
            else:
                return self.minEle
            
    def getMin(self):
        if not self.st:
            return -1
        return self.minEle
    
s = MinStack()
print(s.pop())
print(s.top())
print(s.getMin())
s.push(7)
s.push(3)
s.push(1)
s.push(2)
print(s.top())
s.pop()
print(s.getMin())
s.pop()
print(s.getMin())
s.pop()
print(s.getMin())
```
