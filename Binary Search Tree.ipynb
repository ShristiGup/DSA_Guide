```python
# Floor and Ceil in BST

class Node:
    def __init__(self,val):
        self.key = val
        self.left = None
        self.right = None
        
def insert(root,x):
    if root is None:
        return Node(x)
    if root.key<x:
        root.right = insert(root.right,x)
    elif root.key>x:
        root.left = insert(root.left,x)
    return root

###### Naive sol:- T.C.:- O(n) and S.C:- O(h)

def floor(root,x):           ####### Iterative sol
    res = None               #### T.C.:- O(h)
    while(root):             #### S.C.:- O(1)
        if root.key==x:
            return x
        elif root.key>x:
            root = root.left
        else:
            res = root.key
            root = root.right
    return res

def ceil(root,x):            ####### Iterative sol
    res = None               #### T.C.:- O(h)
    while(root):             #### S.C.:- O(1)
        if root.key==x:
            return x
        elif root.key<x:
            root = root.right
        else:
            res = root.key
            root = root.left
    return res

r = insert(None,50)
r = insert(r, 30)
r = insert(r, 20)
r = insert(r, 40)
r = insert(r, 70)
r = insert(r, 60)
r = insert(r, 80)

print(floor(r,10))
print(floor(r,40))
print(floor(r,58))
print(floor(r,100))

print(ceil(r,10))
print(ceil(r,40))
print(ceil(r,58))
print(ceil(r,81))
```

```python
# Inorder predecessor and successor of given key in BST

class Node:
    def __init__(self,val):
        self.key = val
        self.left = None
        self.right = None
        
def insert(root,x):
    if root is None:
        return Node(x)
    if root.key<x:
        root.right = insert(root.right,x)
    elif root.key>x:
        root.left = insert(root.left,x)
    return root

# Equivalent to floor.............
def pred(root,x):              ######## Iterative sol
    if root is None:           #### T.C.:- O(h)
        return root            #### S.C.:- O(1)
    res = None
    while(root):
        if root.key>x:
            root = root.left
        elif root.key<x:
            res = root.key
            root = root.right
        else:
            if(root.left!=None):
                temp = root.left
                while(temp.right):
                    temp = temp.right
                res = temp.key
            break
    if root is None:
        return None
    return res

# Equivalent to ceil................
def succ(root,x):           ######## Iterative sol
    if root is None:        #### T.C.:- O(h)
        return root         #### S.C.:- O(1)
    res = None
    while(root):
        if root.key<x:
            root = root.right
        elif root.key>x:
            res = root.key
            root = root.left
        else:
            if root.right!=None:
                temp = root.right
                while(temp.left):
                    temp = temp.left
                res = temp.key
            break
    if root is None:
        return None
    return res

        
""" Let us create following BST
              50
           /     \
          30      70
         /  \    /  \
       20   40  60   80
"""
r = insert(None,50)
r = insert(r, 30)
r = insert(r, 20)
r = insert(r, 40)
r = insert(r, 70)
r = insert(r, 60)
r = insert(r, 80)

print(pred(r,80))
print(pred(r,60))
print(pred(r,90))

print(succ(r,80))
print(succ(r,60))
print(succ(r,90))
```

```python
# Find LCA of two nodes in BST 

class Node:
    def __init__(self,val):
        self.key = val
        self.left = None
        self.right = None
        
def insert(root,x):
    if root is None:
        return Node(x)
    if root.key<x:
        root.right = insert(root.right,x)
    elif root.key>x:
        root.left = insert(root.left,x)
    return root

def lca(root,d1,d2):              ######### Recursive sol........
    if root is None:
        return None
    if d1<root.key and d2<root.key:     ### T.C.:- O(h)
        return lca(root.left,d1,d2)     ### S.C.:- O(h)
    elif d1>root.key and d2>root.key:
        return lca(root.right,d1,d2)
    else:
        return root.key

def lca_ite(root,d1,d2):                 ### T.C.:- O(h)
    while(root):                         ### S.C.:- O(1)
        if root.key>d1 and root.key>d2:
            root = root.left
        elif root.key<d1 and root.key<d2:
            root = root.right
        else:
            return root.key

        
""" Let us create following BST
              50
           /     \
          30      70
         /  \    /  \
       20   40  60   80
"""
r = insert(None,50)
r = insert(r, 30)
r = insert(r, 20)
r = insert(r, 40)
r = insert(r, 70)
r = insert(r, 60)
r = insert(r, 80)
print(lca_ite(r,70,80))
```

```python
# Size of the largest BST in a Binary Tree(Hard)

class Node:
    def __init__(self,val):
        self.key = val
        self.left = None
        self.right = None
        
def insert(root,x):
    if root is None:
        return Node(x)
    if root.key<x:
        root.right = insert(root.right,x)
    elif root.key>x:
        root.left = insert(root.left,x)
    return root

INT_MIN = -2**31-1
INT_MAX = 2**31+1

def lar_BST(root):                                        #### Time Complexity: O(N)
    if root is None:                                      #### Auxiliary Space: O(Height of the BST)
        return INT_MAX,INT_MIN,True,None,0
    if root.left==None and root.right==None:
        return root.key,root.key,True,root,1
    
    l = lar_BST(root.left)
    r = lar_BST(root.right)
    
    ret = [0,0,0,0,0]           ##### [min,max,isBST,root,size]
    
    ret[2] = l[2] and r[2] and root.key>l[1] and root.key<r[0]
    ret[0] = min(root.key,l[0],r[0])
    ret[1] = max(root.key,l[1],r[1])
    
    if ret[2]:
        ret[3] = root
        ret[4] = l[4]+r[4]+1
    elif l[4]>r[4]:
        ret[3] = l[3]
        ret[4] = l[4]
    else:
        ret[3] = r[3]
        ret[4] = r[4]
        
    return ret
    

""" Let us create following BST
              50
           /     \
          30      70
         /  \    /  \
       20   40  60   80
"""
r = insert(None,50)
r = insert(r, 30)
r = insert(r, 20)
r = insert(r, 40)
r = insert(r, 70)
r = insert(r, 60)
r = insert(r, 80)
print(lar_BST(r)[4])
```

```python
# Serialize and Deserialize a Binary Tree

class Node:
    def __init__(self,val):
        self.key = val
        self.left = None
        self.right = None
        
def insert(root,x):
    if root is None:
        return Node(x)
    if root.key<x:
        root.right = insert(root.right,x)
    elif root.key>x:
        root.left = insert(root.left,x)
    return root

def serialize(root):                      ### T.C.:- O(N)
    if root is None:                      ### S.C.:- O(N)
        return "X"
    l = serialize(root.left)
    r = serialize(root.right)
    return str(root.key)+","+l+","+r

    """Encodes a tree to a single string.

    :type root: TreeNode
    :rtype: str
    """

def deserialize(data):
    """Decodes your encoded data to tree.

    :type data: str
    :rtype: TreeNode
    """
    arr = data.split(",")
    q = []
    for i in arr:
        q.append(i)
    root = deSer(q)
    return root

def deSer(q):
    if len(q)==0:
        return None
    k = q.pop(0)
    if k=="X":
        return None
    root = Node(int(k))
    root.left = deSer(q)
    root.right = deSer(q)
    return root

def inorder(root):
    if root:
        inorder(root.left)
        print(root.key)
        inorder(root.right)

r = insert(None,50)
r = insert(r, 70)
r = insert(r, 20)
# r = insert(r, 40)
# r = insert(r, 70)
# r = insert(r, 60)
# r = insert(r, 80)
print(serialize(r))
k = deserialize(serialize(r))
inorder(k)
```

```python
# Maximum and Minimum in BST

class Node:
    def __init__(self,val):
        self.key = val
        self.left = None
        self.right = None
        
def insert(root,x):
    if root is None:
        return Node(x)
    if root.key<x:
        root.right = insert(root.right,x)
    elif root.key>x:
        root.left = insert(root.left,x)
    return root

def max_BST(root):               # Time Complexity: O(h) (O(n) in case of skewed tree)
    if root is None:             # Auxiliary Space: O(1)
        return None
    while(root.right):
        root = root.right
    return root.key

def min_BST(root):               # Same as above
    if root is None:
        return None
    while(root.left):
        root = root.left
    return root.key

""" Let us create following BST
              50
           /     \
          30      70
         /  \    /  \
       20   40  60   80
"""
r = insert(None,50)
r = insert(r, 30)
r = insert(r, 20)
r = insert(r, 40)
r = insert(r, 70)
r = insert(r, 60)
r = insert(r, 80)
print(max_BST(r))
print(min_BST(r))
```

```python
# Binary Tree to Binary Search Tree Conversion

class Node:
    def __init__(self,val):
        self.key = val
        self.left = None
        self.right = None
        
def bin_to_BST(root):  ### T.C.:- O(nlogn)
    if root is None:   ### S.C.:- O(n)
        return None
    arr = []
    inorder(root,arr)
    arr.sort()
    i = [0]
    change(root,arr,i)

def change(root,arr,i):
    if root:
        change(root.left,arr,i)
        root.key = arr[i[0]]
        i[0]+=1
        change(root.right,arr,i)
    
def inorder(root,arr):
    if root:
        inorder(root.left,arr)
        arr.append(root.key)
        inorder(root.right,arr)

def nor_inorder(root):
    if root:
        nor_inorder(root.left)
        print(root.key)
        nor_inorder(root.right)
        
root = Node(10)
root.left = Node(30)
root.right = Node(15)
root.left.left = Node(20)
root.right.right = Node(5)
bin_to_BST(root)
nor_inorder(root)
```

```python
# Count BST nodes that lie in a given range

class Node:
    def __init__(self,val):
        self.key = val
        self.left = None
        self.right = None
        
def insert(root,x):
    if root is None:
        return Node(x)
    if root.key<x:
        root.right = insert(root.right,x)
    elif root.key>x:
        root.left = insert(root.left,x)
    return root
        
def count(root,l,r,c):                            ### Time Complexity: O(N)
    if root is None:                              ### Auxiliary Space: O(Height of the BST).
        return 0
    if root.key>=l and root.key<=r:
        c[0]+=1
    if l<root.key:
        count(root.left,l,r,c)
    if r>root.key:
        count(root.right,l,r,c)
        
""" Let us create following BST
              50
           /     \
          30      70
         /  \    /  \
       20   40  60   80
"""
        
r = insert(None,50)
r = insert(r, 30)
r = insert(r, 20)
r = insert(r, 40)
r = insert(r, 70)
r = insert(r, 60)
r = insert(r, 80)
c = [0]
count(r,30,100,c)
print(c[0])
```

```python
# Sorted Array to Balanced BST

class Node:
    def __init__(self,val):
        self.key = val
        self.left = None
        self.right = None

def arrToBST(arr):           ### T.C.:- O(n)
    if not arr:              ### S.C.:- O(h)
        return None
    mid = len(arr)//2
    root = Node(arr[mid])
    root.left = arrToBST(arr[:mid])
    root.right = arrToBST(arr[mid+1:])
    return root

# def arrtoBST(self,nums,i,j):
#     if i>j:
#         return None
#     mid = i+(j-i)//2
#     root = Node(nums[mid])
#     root.left = self.arrtoBST(nums,i,mid-1)
#     root.right = self.arrtoBST(nums,mid+1,j)
#     return root

def preOrder(node):
    if not node:
        return
    print(node.key)
    preOrder(node.left)
    preOrder(node.right)

arr = list(map(int,input().split()))
r = arrToBST(arr)
preOrder(r)
```

```python
# BST to a Tree with sum of all smaller keys

class Node:
    def __init__(self,val):
        self.key = val
        self.left = None
        self.right = None
        
def insert(root,x):
    if root is None:
        return Node(x)
    if root.key<x:
        root.right = insert(root.right,x)
    elif root.key>x:
        root.left = insert(root.left,x)
    return root

def bstToSm(root,s):            ### T.C.:- O(N)
    if root is None:            ### S.C.:- O(h)
        return
    bstToSm(root.left,s)
    s[0]+=root.key
    root.key = s[0]
    bstToSm(root.right,s)
    
def inorder(root):
    if root:
        inorder(root.left)
        print(root.key,end=" ")
        inorder(root.right)
    
""" Let us create following BST
              50
           /     \
          30      70
         /  \    /  \
       20   40  60   80
"""
r = insert(None,50)
r = insert(r, 30)
r = insert(r, 20)
r = insert(r, 40)
r = insert(r, 70)
r = insert(r, 60)
r = insert(r, 80)
s = [0]
inorder(r)
bstToSm(r,s)
print("")
inorder(r)
```

```python
# Merge Two BSTs(Hard)

class Node:
    def __init__(self,val):
        self.key = val
        self.left = None
        self.right = None
        
def insert(root,x):
    if root is None:
        return Node(x)
    if root.key<x:
        root.right = insert(root.right,x)
    elif root.key>x:
        root.left = insert(root.left,x)
    return root

# Naive Method:- T.C.:- O(N1+N2) , S.C.:- O(N1+N2)

def merge_BST(r1,r2):                               ### Optimized Method............  
    st1 = []                                        ### T.C.:- O(N1+N2)
    st2 = []                                        ### S.C.:- O(H1+H2)
    curr1 = r1
    curr2 = r2
    if not r1:
        return inorder(r2)
    if not r2:
        return inorder(r1)
    while(curr1 or st1 or curr2 or st2):
        if curr1 or curr2:
            if curr1:
                st1.append(curr1)
                curr1 = curr1.left
            if curr2:
                st2.append(curr2)
                curr2 = curr2.left
        else:
            if not st1:
                while(st2):
                    curr2 = st2.pop()
                    curr2.left = None
                    inorder(curr2)
                return
            if not st2:
                while(st1):
                    curr1 = st1.pop()
                    curr1.left = None
                    inorder(curr1)
                return
            
            curr1 = st1.pop()
            curr2 = st2.pop()
            
            if curr1.key<curr2.key:
                print(curr1.key)
                curr1 = curr1.right
                st2.append(curr2)
                curr2 = None
            else:
                print(curr2.key)
                curr2 = curr2.right
                st1.append(curr1)
                curr1 = None
                
def inorder(root):
    if root:
        inorder(root.left)
        print(root.key)
        inorder(root.right)
    
""" Let us create following BST
              50
           /     \
          30      70
         /  \    /  \
       20   40  60   80
"""
r1 = insert(None,50)
r1 = insert(r1, 30)
r1 = insert(r1, 20)
r2 = insert(None, 40)
r2 = insert(r2, 70)
r2 = insert(r2, 60)
r2 = insert(r2, 80)
r2 = insert(r2, 90)
r2 = insert(r2, 100)
merge_BST(r2,r1)
```

```python
# Transform a BST to greater sum tree

class Node:
    def __init__(self,val):
        self.key = val
        self.left = None
        self.right = None
        
def insert(root,x):
    if root is None:
        return Node(x)
    if root.key<x:
        root.right = insert(root.right,x)
    elif root.key>x:
        root.left = insert(root.left,x)
    return root

def bstToGr(root,s):            ### T.C.:- O(N)
    if root is None:            ### S.C.:- O(h)
        return
    bstToGr(root.right,s)
    k = root.key
    root.key = s[0]
    s[0]+=k
    bstToGr(root.left,s)
    
def inorder(root):
    if root:
        inorder(root.left)
        print(root.key,end=" ")
        inorder(root.right)
    
""" Let us create following BST
              50
           /     \
          30      70
         /  \    /  \
       20   40  60   80
"""
r = insert(None,11)
r = insert(r, 2)
r = insert(r, 29)
r = insert(r, 1)
r = insert(r, 7)
r = insert(r, 15)
r = insert(r, 40)
r = insert(r, 35)
s = [0]
inorder(r)
bstToGr(r,s)
print("")
inorder(r)
```
