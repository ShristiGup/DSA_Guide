```python
# Occurrences in a sorted array
def count(self,arr, n, x):
    low = 0
    high = n-1
    f = False
    while(low<=high):
        mid = low+(high-low)//2
        if arr[mid]==x:
            high = mid-1
            f = True
        elif arr[mid]<x:
            low = mid+1
        else:
            high = mid-1
    if f==False:
        return 0
    first = high+1
    low = 0
    high = n-1
    while(low<=high):
        mid = low+(high-low)//2
        if arr[mid]==x:
            low = mid+1
        elif arr[mid]<x:
            low = mid+1
        else:
            high = mid-1
    last = low-1
    return last-first+1
```

```python
# Union and Intersection of two sorted arrays
def union(arr1,arr2):
    i = 0
    j = 0
    l = set()
    while(i<len(arr1) and j<len(arr2)):
        if arr1[i]<arr2[j]:
            l.add(arr1[i])
            i+=1
        elif arr2[j]<arr1[i]:
            l.add(arr2[j])
            j+=1
        elif arr1[i]==arr2[j]:
            l.add(arr1[i])
            i+=1
            j+=1
    while(i!=len(arr1)):
        l.add(arr1[i])
        i+=1
    while(j!=len(arr2)):
        l.add(arr2[j])
        j+=1
    return list(l)
    
def intersection(arr1,arr2):
    i = 0
    j = 0
    l = []
    while(i<len(arr1) and j<len(arr2)):
        if arr1[i]==arr2[j]:
            if l and l[-1]==arr1[i]:
                i+=1
                j+=1
            else:
                l.append(arr1[i])
                i+=1
                j+=1
        elif arr1[i]<arr2[j]:
            i+=1
        else:
            j+=1
    if not l:
        return [-1]
    return l

arr1 = list(map(int,input().split()))
arr2 = list(map(int,input().split()))
un = union(arr1,arr2)
inter = intersection(arr1,arr2)
print(*un,sep=" ")
print(*inter,sep=" ")
```

```python
# Implement two stacks in an array
class twoStacks:
    def __init__(self,n):
        self.size = n
        self.arr = [None]*n
        self.top1 = -1
        self.top2 = n
        
    def push1(self,x):
        if self.top1<self.top2-1:
            self.top1 = self.top1+1
            self.arr[self.top1] = x
        else:
            print("Stack Overflow")
            exit(1)
            
    def push2(self,x):
        if self.top1<self.top2-1:
            self.top2 = self.top2-1
            self.arr[self.top2] = x
        else:
            print("Stack Overflow")
            exit(1)
            
    def pop1(self):
        if self.top1>=0:
            x = self.arr[self.top1]
            self.top1 = self.top1-1
            return x
        else:
            print("Stack Underflow")
            exit(1)
    
    def pop2(self):
        if self.top2<self.size:
            x = self.arr[self.top2]
            self.top2 = self.top2+1
            return x
        else:
            print("Stack Underflow")
            exit(1)
            
tS = twoStacks(5)
tS.push1(1)
tS.push2(2)
tS.push1(3)
tS.push2(4)
tS.push2(5)
print(tS.pop1())
tS.push2(70)
tS.push2(60)
print(tS.pop1())
# print(tS.pop1())
print(tS.pop2())
tS.push2(60)
tS.push2(60)
tS.push2(60)
```

```python
# Factorial of very large numbers using arrays
def factorial(self, N):
    res = [None]*4000
    res[0] = 1
    res_size = 1
    i = 2
    while i<=N:
        res_size = self.mul(i,res,res_size)
        i+=1
    ans = []
    for i in range(res_size-1,-1,-1):
        ans.append(res[i])
    return ans

def mul(self,i,res,res_size):
    carry = 0
    j = 0
    while j<res_size:
        prod = res[j]*i+carry
        res[j] = prod%10
        carry = prod//10
        j+=1
    while(carry):
        res[res_size] = carry%10
        carry = carry//10
        res_size+=1
    return res_size
```

```python
# Maximum Product Subarray
def max_prod(arr):
    max_so_far = arr[0]
    prev_max = arr[0]
    prev_min = arr[0]
    curr_max = arr[0]
    for i in range(1,len(arr)):
        curr_max = max(prev_max*arr[i],prev_min*arr[i],arr[i])
        curr_min = min(prev_max*arr[i],prev_min*arr[i],arr[i])
        max_so_far = max(max_so_far,curr_max)
        prev_max = curr_max
        prev_min = curr_min
    return max_so_far
arr = list(map(int,input().split()))
print(max_prod(arr))
```

```python
# Merge Overlapping Intervals

# Naive:- Sort arr and look for merging intervals in the right for each pair(T.C.:- O(NlogN)+O(N^2))
def merge_intervals(arr):
    arr.sort()
    pair = arr[0]                              ### T.C.:- O(NlogN)+O(N)
    ans = []                                   ### S.C.:- O(N) (For storing ans)
    for i in range(1,len(arr)):
        if arr[i][0]<=pair[1]:
            pair[1] = max(arr[i][1],pair[1])
        else:
            ans.append(pair)
            pair = arr[i]
    ans.append(pair)
    return ans
```

```python
# Set Matrix Zeros

# Naive(Brute Force) Method:- Traverse matrix,if 0 is encountered replace non-zero vals in that row and col with a out of range 
# val. T.C.:- O(M*N)*O(M+N) and S.C.:- O(1)

# 2nd Method:- Take two dummy arr,one of size of rows and another cols size. If (i,j)==0,turn same 0,in row and col arr.Again 
# traverse the matrix and make (i,j) 0 if 0 in row or col arr for that index. T.C.:- O(M*N), S.C.:- O(M+N)

def set_zero(mat):                  ### T.c.:- O(M*N)
    col = True                      ### S.C.:- O(1)
    r = len(mat)
    c = len(mat[0])
    for i in range(r):
        if mat[i][0]==0:
            col = False
        for j in range(1,c):
            if mat[i][j]==0:
                mat[i][0]=mat[0][j]=0
    for i in range(r-1,-1,-1):
        for j in range(c-1,0,-1):
            if mat[i][0]==0 or mat[0][j]==0:
                mat[i][j] = 0
        if col==False:
            mat[i][0] = 0
    
m = int(input())
n = int(input())
mat = []
for i in range(m):
    k = list(map(int,input().split()))
    mat.append(k)
set_zero(mat)
for i in mat:
    print(*i,sep=" ")
```

```python
# Next Permutation

def next_per(arr):                 ### T.C.:- O(N)
    n = len(arr)                   ### S.C.:- O(1)
    i = n-2
    while(i>=0 and arr[i]>=arr[i+1]):
        i-=1
    if i>=0:
        j = n-1
        while(j>=0 and arr[i]>=arr[j]):
            j-=1
        arr[i],arr[j] = arr[j],arr[i]
    reverse(arr,i+1,n-1)
    
def reverse(arr,s,e):
    while(s<e):
        arr[s],arr[e] = arr[e],arr[s]
        s+=1
        e-=1
    
arr = list(map(int,input().split()))
next_per(arr)
print(*arr)
```

```python
# Count Inversions

def cnt_inv(arr):
    n = len(arr)
    ans = mergeSort(arr,0,n-1)
    return ans

def mergeSort(arr,l,r):         ### T.C.:- O(NlogN)
    inv_c = 0                   ### S.C.:- O(N)
    if r>l:
        m = l+(r-l)//2
        inv_c+=mergeSort(arr,l,m)
        inv_c+=mergeSort(arr,m+1,r)
        inv_c+=merge(arr,l,m,r)
    return inv_c

def merge(arr,l,m,r):
    inv_c = 0
    n1 = m-l+1
    n2 = r-m
    left = [0]*n1
    right = [0]*n2
    for i in range(n1):
        left[i] = arr[l+i]
    for i in range(n2):
        right[i] = arr[m+1+i]
    i = 0
    j = 0
    k = l
    while(i<n1 and j<n2):
        if left[i]<=right[j]:
            arr[k] = left[i]
            i+=1
            k+=1
        else:
            arr[k] = right[j]
            j+=1
            k+=1
            inv_c+=(m+1-(l+i))
    while(i<n1):
        arr[k] = left[i]
        i+=1
        k+=1
    while(j<n2):
        arr[k] = right[j]
        j+=1
        k+=1
    return inv_c
    
arr = list(map(int,input().split()))
print(cnt_inv(arr))
```

```python
# Best time to buy and sell stock

# Brute:- Look for every pair in right of an index and store max profit, O(N^2) approach
def buy_sell(arr):                      # Type 1: Only 1 transaction allowed
    mn = arr[0]                         ### T.C.:- O(N)
    pr = 0                              ### S.C.:- O(1)
    for i in range(1,len(arr)):
        if arr[i]<mn:
            mn = arr[i]
        elif (arr[i]-mn)>pr:
            pr = arr[i]-mn
    return pr
    
arr = list(map(int,input().split()))
print(buy_sell(arr))
```

```python
# Best time to buy and sell stock

def buy_sell(arr):                      # Type 3: At most 2 transactions(But sell before buying new one)
    n = len(arr)                        ### T.C.:- O(N)
    left = [0]*n                        ### S.C.:- O(N)
    right = [0]*n
    lmin = arr[0]
    rmax = arr[n-1]
    for i in range(1,n):
        if arr[i]<lmin:
            lmin = arr[i]
        left[i] = max(left[i-1],arr[i]-lmin)
    for i in range(n-2,-1,-1):
        if arr[i]>rmax:
            rmax = arr[i]
        right[i] = max(right[i+1],rmax-arr[i])
    pr = right[0]
    for i in range(1,n):
        pr = max(pr,left[i-1]+right[i])
    return pr
    
arr = list(map(int,input().split()))
print(buy_sell(arr))
```

```python
# Pow(x,n)

# Naive:- Mutiply x n times......O(N) time com.(Ask whether given n is positive or negative)
def power(x,n):
    k = n
    if n<0:
        k = -n                           ### T.C.:- O(logN)
    ans = 1                              ### S.C.:- O(1)
    while(k>0):
        if k%2==0:
            x = x*x
            k = k//2
        else:
            ans = ans*x
            k = k-1
    if n<0:
        ans = 1/ans
    return ans
    
x = float(input())
n = int(input())
print(power(x,n))
```

```python
### Grid Unique Paths

# 1st Method:- Recursive sol.......
def grid_path1(m,n,i,j):                 ### T.C.:- Exponential(Probably 2^N)
    if i>=m or j>=n:                     ### S.C.:- Recursion call stack(Exponential,Probably 2^N) 
        return 0
    if i==m-1 and j==n-1:
        return 1
    return grid_path1(m,n,i+1,j)+grid_path1(m,n,i,j+1)

# 2nd Method:- DP sol........
def grid_path2(m,n,i,j,dp):              ### T.C.:- O(M*N)
    if i>=m or j>=n:                     ### S.C.:- O(M*N)(Approx. hash table+recursion call stack)
        return 0
    if i==m-1 and j==n-1:
        return 1
    if dp[i][j]!=-1:
        return dp[i][j]
    else:
        dp[i][j] = grid_path2(m,n,i+1,j,dp)+grid_path2(m,n,i,j+1,dp)
    return dp[i][j]

# 3rd Method:-
def grid_paths(m,n):             
    k = m+n-2                  # Total no of steps
    r = m-1                    # Total right steps or downward steps
    res = 1                    
    for i in range(r):         ### T.C.:- O(m-1) or O(n-1)
        res = res*(k-i)        ### S.C.:- O(1)
        res = res//(i+1)
    return res

m = int(input())
n = int(input())
i = 0
j = 0
global dp
dp = [[-1]*n for i in range(m)]
print(grid_path2(m,n,i,j,dp))
```

```python
# Reverse Pairs(Hard)               
class Solution:
    def reversePairs(self, nums: List[int]) -> int:         ### T.C.:- O(NlogN)+O(N)+O(N)(Merge sort+merge op.+counting op.)
        n = len(nums)                                       ### S.C.:- O(N)
        ans = self.mergeSort(nums,0,n-1)
        return ans

    def mergeSort(self,nums,l,r):
        cnt = 0
        if r>l:
            m = l+(r-l)//2
            cnt+=self.mergeSort(nums,l,m)
            cnt+=self.mergeSort(nums,m+1,r)
            cnt+=self.merge(nums,l,m,r)
        return cnt

    def merge(self,nums,l,m,r):
        cnt = 0
        n1 = m-l+1
        n2 = r-m
        j = m+1
        for i in range(l,m+1):
            while(j<=r and nums[i]>2*nums[j]):
                j+=1
            cnt+=(j-(m+1))
        left = [0]*n1
        right = [0]*n2
        for i in range(n1):
            left[i] = nums[i+l]
        for i in range(n2):
            right[i] = nums[m+1+i]
        i = 0
        j = 0
        k = l
        while(i<n1 and j<n2):
            if left[i]<=right[j]:
                nums[k] = left[i]
                k+=1
                i+=1
            else:
                nums[k] = right[j]
                j+=1
                k+=1
        while(i<n1):
            nums[k] = left[i]
            i+=1
            k+=1
        while(j<n2):
            nums[k] = right[j]
            j+=1
            k+=1
        return cnt
```

```python
# 3 sum
def threeSum(self, nums: List[int]) -> List[List[int]]:
    ans = []
    nums.sort()
    n = len(nums)
    for i in range(0,n-2):
        if i==0 or (i>0 and nums[i]!=nums[i-1]):
            lo = i+1
            hi = n-1
            s = 0-nums[i]
            while(lo<hi):
                if(nums[lo]+nums[hi]==s):
                    ans.append([nums[i],nums[lo],nums[hi]])
                    while(lo<hi and nums[lo]==nums[lo+1]):
                        lo+=1
                    while(lo<hi and nums[hi]==nums[hi-1]):
                        hi-=1
                    lo+=1
                    hi-=1
                elif nums[lo]+nums[hi]<s:
                    lo+=1
                else:
                    hi-=1
    return ans      
```

```python
# Longest Consecutive Sequence

# Naive sol:- Sort the array and look for lar consecutive seq. T.C.:- O(nlogn),S.C.:- O(1)

def long_seq(arr):                                 ### T.C.:- O(N)
    s = set(arr)                                   ### S.C.:- O(N)
    mx = 0
    for i in arr:
        if i-1 not in s:
            num = i
            curr = 1
            while(num+1 in s):
                num+=1
                curr+=1
            if curr>mx:
                mx = curr
    return mx
    
arr = list(map(int,input().split()))
print(long_seq(arr))
```

```python
# Longest sub-array having sum k

def long_subarr(arr,k):            # Time Complexity: O(n)
    d = dict()                     # Auxiliary Space: O(n)
    curr = 0
    mx = 0
    for i in range(len(arr)):
        curr+=arr[i]
        if curr==k:
            mx = i+1
        elif curr-k in d:
            mx = max(mx,i-d[curr-k])
        if curr not in d:
            d[curr] = i
    return mx
        
arr = list(map(int,input().split()))
k = int(input())
print(long_subarr(arr,k))
```

```python
# Largest Subarray with 0 sum

def maxLen(self, n, arr):  # T.C.:- O(N)
    d = dict()             # S.C.:- O(N)
    curr = 0
    mx = 0
    for i in range(n):
        curr+=arr[i]
        if curr==0:
            mx = i+1
        elif curr in d:
            mx = max(mx,i-d[curr])
        if curr not in d:
            d[curr] = i
    return mx
```

```python
# Largest subarray of 0's and 1's

def maxLen(self,arr, N):      # T.C.:- O(N)
    for i in range(N):        # S.C.:- O(N)
        if not arr[i]:
            arr[i] = -1
    d = {}
    mx = 0
    curr = 0
    for i in range(N):
        curr+=arr[i]
        if curr==0:
            mx = i+1
        elif curr in d:
            mx = max(mx,i-d[curr])
        if curr not in d:
            d[curr] = i
    return mx
```

```python
# Count number of subarrays with given XOR

def sub_xor(arr,k):               ### T.C.:- O(NlogN)/O(N)
    d = dict()                    ### S.C.:- O(N)
    cnt = 0
    xor = 0
    for i in arr:
        xor = xor^i
        if xor==k:
            cnt+=1
        y = xor^k
        if d.get(y)!=None:
            cnt+=d.get(y)
        if xor in d:
            d[xor]+=1
        else:
            d[xor] = 1
    
arr = list(map(int,input().split()))
k = int(input())
print(sub_xor(arr,k))
```

```python
# Remove Duplicate from Sorted array

def rem_dup(arr):                       ### T.C.:- O(N)
    i = 0                               ### S.C.:- O(1)
    j = 0
    n = len(arr)
    while(j<n):
        if arr[i]!=arr[j]:
            i+=1
            arr[i] = arr[j]
        j+=1
    return i+1
    
arr = list(map(int,input().split()))
print(rem_dup(arr))
```

```python
# Max consecutive ones

def max_one(arr):             ### T.C.:- O(N)
    cnt = 0                   ### S.C.:- O(1)
    max_cnt = 0
    for i in arr:
        if i:
            cnt+=1
            max_cnt = max(max_cnt,cnt)
        else:
            cnt = 0 
    return max_cnt
    
arr = list(map(int,input().split()))
print(max_one(arr))
```

```python
# 4 sum

def fourSum(self, arr, k):                      ### T.C.:- O(N^3)
    arr.sort()                                  ### S.C.:- O(1)
    n = len(arr)
    s = []
    i = 0
    while(i<n):
        j = i+1
        while(j<n):
            rem = k-arr[i]-arr[j]
            left = j+1
            right = n-1
            while(left<right):
                t = arr[left]+arr[right]
                if t<rem:
                    left+=1
                elif t>rem:
                    right-=1
                else:
                    ans = [arr[i],arr[j],arr[left],arr[right]]
                    s.append(ans)
                    while(left<right and arr[left]==ans[2]):
                        left+=1
                    while(left<right and arr[right]==arr[3]):
                        right-=1
            while(j+1<n and arr[j]==arr[j+1]):
                j+=1
            j+=1
        while(i+1<n and arr[i]==arr[i+1]):
            i+=1
        i+=1
    return s
```
