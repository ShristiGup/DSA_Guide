```python
## Complete LinkedList example containing printlist,push,palindrome,move last to front,swap nodes,pairwise swap nodes.
class LinkedList:
    def __init__(self):
        self.head = None
        
    def printList(self):
        temp = self.head
        while(temp):
            print(temp.data,end=" ")
            temp = temp.next
        print("")
            
    def push(self,x):
        new_n = Node(x)
        new_n.next = self.head
        self.head = new_n
        
## Using Stack.............    
    def palin(self):
        s = []
        curr = self.head
        while(curr):
            s.append(curr.data)
            curr = curr.next
        curr = self.head
        while(curr):
            k = s.pop()
            if k==curr.data:
                curr = curr.next
            else:
                return False
        return True
        
    def makelastfirst(self):
        curr = self.head
        if curr==None or curr.next==None:
            return
        while(curr.next):
            prev = curr
            curr = curr.next
        prev.next = None
        curr.next = self.head
        self.head = curr
    
    def swapNodes(self,key1,key2):
        prev1 = None
        prev2 = None
        curr1 = self.head
        curr2 = self.head
        while curr1 and curr1.data!=key1:
            prev1 = curr1
            curr1 = curr1.next
        while curr2 and curr2.data!=key2:
            prev2 = curr2
            curr2 = curr2.next
        if curr1==None or curr2==None:
            return
        if prev1!=None:
            prev1.next = curr2
        else:
            self.head = curr2
        if prev2!=None:
            prev2.next = curr1
        else:
            self.head = curr1
        temp = curr1.next
        curr1.next = curr2.next
        curr2.next = temp
        return
    
    def pairWiseSwap(self):
        curr = self.head
        if curr.next==None:
            return                                     ###################################################
        prev = None                                    # VERY IMPORTANT
        while(curr):
            n1 = curr
            n2 = curr.next
            if n2==None:
                return
            if prev==None:
                self.head = n2
            else:
                prev.next = n2
            prev = curr
            n1.next = n2.next
            n2.next = prev
            curr = curr.next
        return
        
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

llist = LinkedList()
llist.push(4)
llist.push(6)
llist.push(3)
llist.push(7)
llist.push(2)
llist.printList()
# print(llist.palin())
# llist.makelastfirst()
# llist.swapNodes(2,7)
llist.pairWiseSwap()
llist.printList()
```

```python
# Intersection Point in Y Shaped Linked Lists

# 1st Optimal
def intersetPoint(head1,head2):          ### T.C.:- O(2M) ~~ O(M)(Length of larger ll)
    a = head1                            ### S.C.:- O(1)
    b = head2
    count1 = 0
    count2 = 0
    while a:
        count1+=1
        a = a.next
    while b:
        count2+=1
        b = b.next
    if count1>count2:
        d = count1-count2
        bg = head1
        sm = head2
    else:
        d = count2-count1
        bg = head2
        sm = head1
    for i in range(0,d):
        bg = bg.next
    while bg and sm:
        if bg == sm:
            return bg.data
        bg = bg.next
        sm = sm.next
    return -1

# 2nd Optimal
def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
    if headA is None or headB is None:                ### T.C.:- O(2M) ~~ O(M)(Length of larger ll)
        return None                                   ### S.C.:- O(1)
    curr1 = headA
    curr2 = headB
    while(curr1!=curr2):
        if curr1:
            curr1 = curr1.next
        else:
            curr1 = headB
        if curr2:
            curr2 = curr2.next
        else:
            curr2 = headA
    return curr1
```

```python
# Reverse a Linked List

class LinkedList:
    def __init__(self):
        self.head = None
    
    def push(self,x):
        newNode = Node(x)
        newNode.next = self.head
        self.head = newNode
        
    def printList(self):
        temp = self.head
        while(temp):
            print(temp.data,end=" ")
            temp = temp.next
    
    def reverse_ll(self):             # Iterative.....
        prev = None                   # O(N) time
        curr = self.head              # O(1) space
        while(curr!=None):
            next_n = curr.next
            curr.next = prev
            prev = curr
            curr = next_n
        self.head = prev
    
    def reverse_ll(self,curr,prev):     # Recursive.....
        if curr is None:                # O(N) time
            self.head = prev            # O(1) space
            return
        next_n = curr.next
        curr.next = prev
        return self.reverse_ll(next_n,curr)

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

ll = LinkedList()
ll.push(1)
ll.push(2)
ll.push(3)
ll.push(4)
ll.push(5)
ll.printList()
# ll.reverse_ll()
ll.reverse_ll(ll.head,None)
print("")
ll.printList()
```

```python
# Rotate a LinkedList

class LinkedList:
    def __init__(self):
        self.head = None
    
    def push(self,x):
        newNode = Node(x)
        newNode.next = self.head
        self.head = newNode
        
def printList(head):
    temp = head
    while(temp):
        print(temp.data,end=" ")
        temp = temp.next

# Naive:- Traverse k times through whole ll and make last ele first or vice versa......O(N*k) method
def rotate_ll(head,k):                          ### T.C.:- O(N)
    if not head or not head.next or k==0:       ### S.C.:- O(1)
        return head
    l = 1
    curr = head
    while(curr.next):
        l+=1
        curr = curr.next
    curr.next = head
    k = k%l
    k = l-k                                ### Right shift happens here(Eliminate this step for left shift)
    for i in range(k):
        curr = curr.next
    head = curr.next
    curr.next = None
    return head
        
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

ll1 = LinkedList()
ll1.push(5)
ll1.push(4)
ll1.push(3)
ll1.push(2)
ll1.push(1)
ll1.push(0)
h = rotate_ll(ll1.head,1)
printList(h)
```

```python
# Detect a cycle in Linked List

def det_cyc(head):                      ### T.C.:- O(N)
    slow = head                         ### S.C.:- O(1)
    fast = head
    while(fast and fast.next):
        slow = slow.next
        fast = fast.next.next
        if slow==fast:
            return True
    return False
```

```python
# Find the starting point of the Loop of LinkedList

def st_pt(head):                              ### T.C.:- O(N)
    if head==None or head.next==None:         ### S.C.:- O(1)
        return None
    slow = head
    fast = head
    while(fast and fast.next):
        slow = slow.next
        fast = fast.next.next
        if slow==fast:
            break
    if slow!=fast:
        return None
    slow = head
    while(slow!=fast):
        slow = slow.next
        fast = fast.next
    return slow
```
