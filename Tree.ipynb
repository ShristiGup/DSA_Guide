```python
# Inorder, Preorder and Postorder Traversal(Recursive sol.)
class Node:
    def __init__(self,key):
        self.left = None
        self.right = None
        self.val = key
        
def inorder(root):
    if root:
        inorder(root.left)
        print(root.val,end=" ")
        inorder(root.right)

def preorder(root):
    if root:
        print(root.val,end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val,end=" ")
    
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
print("Inorder:")
inorder(root)
print("\nPreorder")
preorder(root)
print("\nPostorder")
postorder(root)

## Time Complexity: O(N) or theta(N)
## Aux. Space: O(h+1) = O(h) or theta(h), h = height of tree(+1 for null)
```

```python
# Get level of a node in binary tree
class Node:
    def __init__(self,key):
        self.val = key
        self.left = None
        self.right = None
        
# Iterative sol...............       
def levelOfNode(root,key):
    q = [root]
    q.append(None)
    level = 1
    while(len(q)>0):
        temp = q.pop(0)
        if temp==None:
            if len(q)==0:
                return -1
            if len(q)!=0:
                q.append(None)
            level+=1
        else:
            if temp.val == key:
                return level
            if temp.left:
                q.append(temp.left)
            if temp.right:
                q.append(temp.right)
                
# Recursive sol............            
def levelOfNode(root,key,level=1):
    if root is None:
        return 0
    if root.val==key:
        return level
    l = levelOfNode(root.left,key,level+1)
    if l!=0:
        return l
    l = levelOfNode(root.right,key,level+1)
    return l
    
root = Node(10)
root.left = Node(20)
root.right = Node(70)
root.left.left = Node(40)
root.left.right = Node(70)
root.right.left = Node(90)
root.right.right = Node(80)
print(levelOfNode(root,70))
```

```python
# Reverse Level Order Traversal of a Binary Tree

class Node:
    def __init__(self,key):
        self.val = key
        self.left = None
        self.right = None
        
def revLevel(root):
    if root is None:
        return
    st = []
    q = [root]
    while(len(q)>0):
        k = q.pop(0)
        if k.right:
            q.append(k.right)
        if k.left:
            q.append(k.left)
        st.append(k.val)
    while(len(st)>0):
        k = st.pop()
        print(k,end=" ")

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
revLevel(root)
```

```python
# Build Tree from Preorder and Inorder(Naive: O(n^2))
# Time Complexity : O(n)
# Space Complexity : O(n)

class Node:
    def __init__(self,key):
        self.val = key
        self.left = None
        self.right = None
        
def buildTree(preorder,inorder):
    global mp
    mp = {}
    for i in range(len(preorder)):
        mp[inorder[i]] = i
    in_st = 0
    in_end = len(inorder)-1
    start = [0]
    root = newTree(preorder,inorder,start,in_st,in_end)
    return root

def newTree(preorder,inorder,start,in_st,in_end):
    if in_st>in_end:
        return None
    curr = preorder[start[0]]
    root = Node(curr)
    start[0]+=1

    if in_st==in_end:
        return root

    ind = mp[curr]

    root.left = newTree(preorder,inorder,start,in_st,ind-1)
    root.right = newTree(preorder,inorder,start,ind+1,in_end)

    return root

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val,end=" ")
        
inorder = list(map(int,input().split()))
preorder = list(map(int,input().split()))
r = buildTree(preorder,inorder)
postorder(r)
```

```python
# Build Tree from Postorder and Inorder(Naive: O(n^2))
# Time Complexity : O(n)
# Space Complexity : O(n)

class Node:
    def __init__(self,key):
        self.val = key
        self.left = None
        self.right = None
        
def buildTree(In, post, n):
    global mp
    mp = {}
    for i in range(n):
        mp[In[i]] = i
    start = [n-1]
    in_st = 0
    in_end = n-1
    root = newTree(In,post,start,in_st,in_end)
    return root
    
def newTree(In,post,start,in_st,in_end):
    if in_st>in_end:
        return None
    curr = post[start[0]]
    root = Node(curr)
    start[0]-=1
    
    if in_st==in_end:
        return root
    
    ind = mp[curr]
    
    root.right = newTree(In,post,start,ind+1,in_end)
    root.left = newTree(In,post,start,in_st,ind-1)
    
    return root

def preorder(root):
    if root:
        print(root.val,end=" ")
        preorder(root.left)
        preorder(root.right)
        
inorder = list(map(int,input().split()))
postorder = list(map(int,input().split()))
n = len(inorder)
r = buildTree(inorder,postorder,n)
preorder(r)
```

```python
# Build Tree from Level order and Inorder
# Time Complexity : O(n^2)
# Space Complexity : O(n)

class Node:
    def __init__(self,key):
        self.val = key
        self.left = None
        self.right = None
        
def buildTree(ino, level, n):
    global mp
    mp = {}
    for i in range(len(level)):
        mp[level[i]] = i
    in_st = 0
    in_end = len(level)-1
    root = newTree(level,ino,in_st,in_end)
    return root
    
def newTree(level,ino,in_st,in_end):
    if in_st>in_end:
        return None
    
    ind = in_st
    for j in range(in_st+1,in_end+1):
        if mp[level[ino[j]]]<mp[level[ino[ind]]]:
            ind = j
    
    root = Node(ino[ind])
    root.left = newTree(level,ino,in_st,ind-1)
    root.right = newTree(level,ino,ind+1,in_end)
    
    return root

def preorder(root):
    if root:
        print(root.val,end=" ")
        preorder(root.left)
        preorder(root.right)
        
ino = list(map(int,input().split()))
level = list(map(int,input().split()))
n = len(ino)
r = buildTree(ino,level,n)
preorder(r)
```
